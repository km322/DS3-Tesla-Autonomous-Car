```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import os
```

```python
current_dir = os.getcwd()
print(current_dir)
```

```python
RGB_dataA_files = sorted(os.listdir(current_dir + "/archive-2/dataA/dataA/CameraRGB"))[:5]
Seg_dataA_files = sorted(os.listdir(current_dir + "/archive-2/dataA/dataA/CameraSeg"))[:5]

for i in range(len(RGB_dataA_files)):
    RGB_dataA_files[i] = cv2.imread(current_dir + "/archive-2/dataA/dataA/CameraRGB/" + RGB_dataA_files[i])
    Seg_dataA_files[i] = cv2.imread(current_dir + "/archive-2/dataA/dataA/CameraRGB/" + Seg_dataA_files[i], cv2.IMREAD_GRAYSCALE)
    

fig, axes = plt.subplots(len(RGB_dataA_files), 2, figsize=(20, 30))

for i in range(len(RGB_dataA_files)):
    axes[i][0].imshow(RGB_dataA_files[i])
    axes[i][1].imshow(Seg_dataA_files[i])

plt.show()
```
